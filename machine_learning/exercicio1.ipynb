```python
# dataset generator (script version)
import numpy as np
N=200
xscale=10
#noise
mean=0
stdev=1
x=np.random.rand(N)*xscale
noise=np.random.normal(loc=mean,scale=stdev,size=N)
y=x*2+5+noise

```

```python


# dataset generator function

def datgen(N=100,xscale=1,noise_mean=0,noise_stdev=1):
    x=np.random.rand(N)*xscale
    noise=np.random.normal(loc=noise_mean,scale=noise_stdev,size=N)
    y=x*2+5+noise
    return (x,y)

```

```python


import matplotlib.pyplot as plt

(x,y)=datgen()
plt.scatter(x,y)

```

```python
# sandbox 
# run with different values for N
# measure standard deviation of coefficients

from sklearn.linear_model import LinearRegression

(x,y)=datgen(2000,1,noise_stdev=0.3)
X=x[:,np.newaxis]

model=LinearRegression()
model.fit(X,y)
model.coef_[0],model.intercept_, model.score(X,y)
```

```python
# plot coefficient estimates with N

Repeats=5
Ns=np.arange(10,5000,step=10)
coeffs=[]
for N in Ns:
    (x,y)=datgen(N,1,noise_stdev=0.5)
    X=x[:,np.newaxis]
    this_coeff=[]
    for k in range(Repeats):
        model=LinearRegression()
        model.fit(X,y)
        this_coeff.append(model.coef_)
    coeffs.append(np.mean(this_coeff))
coeffs=np.array(coeffs)
plt.scatter(Ns,coeffs); #semicolon avoids text output

```

```python
# plot intercept estimates with N
# Programming exercises: 
#    1. adapt the code to do both coefficient and intercept
#    2. define a function that does these tasks


Repeats=5
Ns=np.arange(10,5000,step=10)
coeffs=[]
for N in Ns:
    (x,y)=datgen(N,1,noise_stdev=0.5)
    X=x[:,np.newaxis]
    this_coeff=[]
    for k in range(Repeats):
        model=LinearRegression()
        model.fit(X,y)
        this_coeff.append(model.intercept_)
    coeffs.append(np.mean(this_coeff))
coeffs=np.array(coeffs)
plt.scatter(Ns,coeffs); #semicolon avoids text output

```

```python
# plot R2 estimates with N
# Data science experiment: try with different noise_stdev

Repeats=5
Ns=np.arange(10,5000,step=10)
scores=[]
for N in Ns:
    (x,y)=datgen(N,1,noise_stdev=1)
    X=x[:,np.newaxis]
    this_score=[]
    for k in range(Repeats):
        model=LinearRegression()
        model.fit(X,y)
        this_score.append(model.score(X,y))
    scores.append(np.mean(this_score))
scores=np.array(scores)
plt.scatter(Ns,scores); #semicolon avoids text output
```

```python
# plot R2 estimates with dispersion
# 

Repeats=5
N=100
Disps=np.arange(0,5,step=0.01)
scores=[]
for D in Disps:
    (x,y)=datgen(N,1,noise_stdev=D)
    X=x[:,np.newaxis]
    this_score=[]
    for k in range(Repeats):
        model=LinearRegression()
        model.fit(X,y)
        this_score.append(model.score(X,y))
    scores.append(np.mean(this_score))
scores=np.array(scores)
plt.scatter(Disps,scores); #semicolon avoids text output
```
